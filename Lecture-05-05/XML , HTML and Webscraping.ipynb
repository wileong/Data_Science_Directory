```python
#exercise: count up how many works ea conductor conducted. Who conducted the most works?



#data.find_all("work")[0].find_all("composerName")

conductor_counts = {}

for conductorTag in data.find_all("conductorName"):
    
    conductorName = conductorTag.string #multiple names
    for conductor in conductorName.split(";  "):
            
            if conductorName in conductor_counts:
        
                conductor_counts[conductorName] += 1
        
            else:
        
                conductor_counts[conductorName] = 1
    
    
        
```

```python
import pandas as pd
pd.Series(conductor_counts).sort_values()
```

```python
#also: how many programs featured at least one work by betthoven?

beethoven_counts = 0


for p in data.find_all("program"):
    for c in p.find_all("composerName"):
        if "Beethoven" in c.string:
            beethoven_counts += 1
            break # at least one beethoven
    
    
```
