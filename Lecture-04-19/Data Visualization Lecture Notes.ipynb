```python
import numpy as np

x = np.linspace(0,10,30)
# spaced by 30 anf 0 = min 10 = max
y = np.sin(x)
```

```python
plt.plot(y, 'ko') #'k' is black
```

```python
plt.plot(y, 'ko-')
#add '-' to connect dots with lines
```

```python
plt.plot(y, '-o', linewidth = 5, markersize = 15)
```

```python
plt.plot(y, '-o', linewidth = 5, markersize = 15, alpha = 0.5)
#alpha is transparency; default = 1
```

```python
#call plot multiple times within SAME CELL to plot on top of ea other

z = np.cos(x)

plt.plot(x,y, 'ko-')
plt.plot(x,z)

#calling plt.figure() before a new plot will create a sep graph
```

```python
import pandas as pd
data = pd.read_csv("/data/harris.csv")
```

```python
data.plot.hist()
# must call .plot first
```

```python
# Make a histogram of the beginning salaries of the employees

data['Bsal'].plot.hist(alpha = .4) # on a series
```

```python
#Make a scatterplot to understnad the relation between exp and beginning salary
# we need a dataframe for two vars

data.plot.scatter(x = 'Exper', y ='Bsal')
#relationship: at first, little exp = low bsal, but as time goes by, it's hard to tell where bsal will be
```

```python
data.plot.hist(alpha = .4)
# plots every quant var on histogram.
# this is bad
```

```python
data[ ['Bsal', 'Sal77'] ].plot.hist(alpha = .4)
# pass it a dataframe to plot sep vars
```

```python
# plot histograms of male and female beginning salaries on the same figure.

data.ix[ data['Sex'] == 'Male', 'Bsal' ].plot.hist(alpha = .4)
data.ix[ data['Sex'] == 'Female', 'Bsal' ].plot.hist(alpha = .2)
plt.legend(['Male','Female']) # pass in labels in order or plotted, but MUST IMPORT from matplotlib first!
```

```python
# bar graphs vs histograms:

#bar graphs graph categorical vars like female or male
# histograms are for QUANT vars
```

```python
#data.sex.plot.bar() doesn't work!

data.groupby('Sex').Bsal.mean()
```

```python
#Female and Male are indicies; groupby gets series into this form
```

```python
data.groupby('Sex').Bsal.mean().plot.bar()
```

```python
bsal = data.groupby('Sex').Bsal
bsal.mean().plot.bar(yerr = 1.96 * bsal.std()/ bsal.count() ** .5)
# height of bar(x-bar) += t.ppf(.975) * s/sqrt(n) ; right side is yerr
#getting standard error bars
# notice std error bar for males is smaller because n for males is smaller
```

```python
# Making Matplotlib Prettier ========================================================================
```

```python
with plt.style.context("fivethirtyeight"): 
    data.ix[ data['Sex'] == 'Male', 'Bsal' ].plot.hist(alpha = .4)
    data.ix[ data['Sex'] == 'Female', 'Bsal' ].plot.hist(alpha = .2)
    plt.legend(['Male','Female']) # pass in labels in order or plotted, but MUST IMPORT from matplotlib first!
```

```python
with plt.style.context("ggplot"): #use grayscale if for public view
    data.ix[ data['Sex'] == 'Male', 'Bsal' ].plot.hist(alpha = .4)
    data.ix[ data['Sex'] == 'Female', 'Bsal' ].plot.hist(alpha = .2)
    plt.legend(['Male','Female']) # pass in labels in order or plotted, but MUST IMPORT from matplotlib first!
```
