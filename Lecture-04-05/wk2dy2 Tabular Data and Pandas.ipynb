```python
#In-class: Compute a 95% CI for avg begin salary of male employees. Then compute a 95% CI for beginning
#of salary of female employees. What do you conclude?

#can look up quantiles of the distributions in pack

from scipy.stats import t

m = data.ix[ data["Sex"] == 'Male','Bsal' ]
print(m.mean() - t.ppf(.975, m.count() - 1) * m.std()/ np.sqrt(m.count()) )
print(m.mean() + t.ppf(.975, m.count() - 1) * m.std()/ np.sqrt(m.count())  )
```

```python
f = data.ix[ data["Sex"] == 'Female','Bsal' ]
print(f.mean() - t.ppf(.975, f.count() - 1) * f.std()/ np.sqrt(f.count()) )
print(f.mean() + t.ppf(.975, f.count() - 1) * f.std()/ np.sqrt(f.count()) )
```
