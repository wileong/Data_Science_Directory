Calculate the number of funded companies and the number of total companies, sliced by industry and gender. To make it a bit easier to compare numbers for men and women, let's pivot gender from rows to columns, so that we can see the percentages for men and women side-by-side. Store this pivoted table in a variable; you'll need it for the next part.

```python
# YOUR CODE HERE
#raise NotImplementedError()



total_companies = data.groupby(["Industry", "Entrepreneur Gender"]).count()
#companies_funded = data.ix[data['Deal'] == 'Yes'].groupby(["Industry", "Entrepreneur Gender"]).count()
#companies_funded = companies_funded.reset_index()
#companies_funded

def get_percentage(series):
    
    total = series.count()
    
    return( (series[series == 'Yes'].count()/total  ) )

companies_funded = data.groupby(['Industry', 'Entrepreneur Gender'])['Deal'].agg([get_percentage])
companies_funded = companies_funded.reset_index() #<---this is necesary!

data_pivoted = companies_funded.pivot_table( index = 'Industry', columns = 'Entrepreneur Gender')
data_pivoted = data_pivoted.fillna(0)
data_pivoted
```

Use your table to compute 95% confidence intervals for the difference between the percentage of funded companies for men and women, one interval for each industry. Are there any industries where the difference is especially stark and statistically significant?

**HINT**: If it's been a while since your last statistics course, you are looking for a confidence interval for the difference of two proportions. Make sure you used the "pooled" estimate in calculating the standard error.

```python
# YOUR CODE HERE
#raise NotImplementedError()

#def lower(x):
#   n = x.count()
#   return x.mean() - t.ppf(.975,n) * x.std()/ np.sqrt(n)
    
#def upper(x):
#    n = x.count()
#    return x.mean() + t.ppf(.975, n) * x.std()/ np.sqrt(n)

# pooled proportion +- z * Standard Error

#(p1 - p2) +- zscore * (sqrt( ( (p1*(1-p1))/n1 ) + ( ( p2*(1-p2 ) )/ n2  ) ) )
# where p1&p2 are proportions of population , n1 & n2 are sample sizes of population


from scipy.stats import norm
import numpy as np

#total_females_in_industry = data.ix[data['Entrepreneur Gender'] == 'Female' ,"Industry"].count()
#total_males_in_industry = data.ix[ data['Entrepreneur Gender'] == 'Male', 'Industry' ].count()

data_pivoted['% Diff Funded for Gender'] = data_pivoted['get_percentage']['Female'] - data_pivoted['get_percentage']['Male']


# Translation: get total males/females per industry by imposing condition that Gender = male or female, then
#group by row industry and column Gender and finally counting total entries that fit these conditions
data_pivoted['Total females per industry'] = data.ix[data['Entrepreneur Gender'] == 'Female'].\
groupby(['Industry'])['Entrepreneur Gender'].count()
data_pivoted['Total females per industry'] = data_pivoted['Total females per industry'].fillna(0)

data_pivoted['Total males per industry'] = data.ix[data['Entrepreneur Gender'] == 'Male'].\
groupby(['Industry'])['Entrepreneur Gender'].count()



p1 = data_pivoted[ ('get_percentage', 'Female') ]
n1 = data_pivoted[ 'Total females per industry' ]
p2 = data_pivoted[ ('get_percentage', 'Male') ]
n2 = data_pivoted[ 'Total males per industry' ]

#data_pivoted['random col'] =  ( (p1 - p1**2 ) /n1 ) + ( (p2- p2**2 ) /n2 )

#Calculating standard error:
data_pivoted['Margin of Error'] = norm.ppf(.975) * np.sqrt( ( (p1 - p1**2 ) /n1 ) + ( (p2- p2**2 ) /n2 )   ) 

data_pivoted['lower'] = data_pivoted['% Diff Funded for Gender'] - data_pivoted['Margin of Error']

data_pivoted['upper'] = data_pivoted['% Diff Funded for Gender'] + data_pivoted['Margin of Error']

data_pivoted

#data_pivoted['lower'] = data_pivoted['% Diff Funded for Gender'] - norm.ppf(.975) * np.sqrt(  )
```
