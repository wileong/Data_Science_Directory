Notice that Amount and Equity are currently being treated as categorical variables (dtype: object). This is because they contain non-numerical characters, like $ and %, so Pandas assumes they are strings. Use [Pandas' string processing capabilities](http://pandas.pydata.org/pandas-docs/stable/text.html) to strip these characters. Then, convert these columns to quantitative variables (i.e., with a dtype of int or float).

```python
# YOUR CODE HERE
#raise NotImplementedError()
"""
#only strips from begin or end...
data.Amount = data.Amount.str.strip(',') # doesn't work

"""
data.Amount = data.Amount.str.replace(',', '')

data.Amount = data.Amount.str.lstrip('$')
data.Equity = data.Equity.str.rstrip('%')

#fill NaN's first...
data['Amount'] = data['Amount'].fillna(0)
data['Equity'] = data['Equity'].fillna(0)

data['Amount'] = data['Amount'].replace('n/a', '')
data['Equity'] = data['Equity'].replace('n/a', '')

#convert categorical var 'Amount' and 'Equity' to a quantitative var; 
#currently dtype of Amount and Equity is Object, check with data.dtypes


data.Amount = pd.to_numeric(data.Amount)
#data.dtypes

#data.Equity = data.Equity.astype(float)
#Strangely, Strings in Equity will not convert to float... so alternatively,

data.Equity = pd.to_numeric(data.Equity)
#data.dtypes


data

```

What percentage of companies led by male entrepreneurs were funded? What percentage of companies led by female entrepreneurs were funded? Report 95% confidence intervals for each.

```python
# YOUR CODE HERE
#raise NotImplementedError()
from scipy.stats import norm
import numpy as np



m_total = data.ix[ (data["Entrepreneur Gender"] == "Male")  , "Deal" ].count()
f_total = data.ix[ (data["Entrepreneur Gender"] == "Female")  , "Deal" ].count()



"""
print(m_total)
print(f_total)
"""

m_funded = data.ix[(data["Entrepreneur Gender"]  == 'Male') & (data["Deal"] == 'Yes') , "Deal"].count()

f_funded = data.ix[(data["Entrepreneur Gender"]  == 'Female') & (data["Deal"] == 'Yes') , "Deal"].count()

"""
print(m_funded)
print(f_funded)
"""

m_funded_percentage = (m_funded/m_total) 
f_funded_percentage = (f_funded/f_total)

print(m_funded_percentage)
print(f_funded_percentage)


m_lower_bound = m_funded_percentage - norm.ppf(.975) * \
    np.sqrt( (m_funded_percentage - m_funded_percentage**2) /m_total)

m_upper_bound = m_funded_percentage + norm.ppf(.975) * \
    np.sqrt( (m_funded_percentage - m_funded_percentage**2 )/m_total) 

f_lower_bound = f_funded_percentage - norm.ppf(.975) * \
    np.sqrt( (f_funded_percentage - f_funded_percentage**2) /f_total) 

f_upper_bound = f_funded_percentage + norm.ppf(.975) *\
    np.sqrt( (f_funded_percentage - f_funded_percentage**2) /f_total)  

print(m_lower_bound)
print(m_upper_bound)
print(f_lower_bound)
print(f_upper_bound)
```
