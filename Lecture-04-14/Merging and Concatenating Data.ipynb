```python
data_list = []

for filename in filenames:
    
    d = pd.read_csv("/data/babynames/" + filename, names = ["Name" ,"Gender", "Count" + filename[3:7]]) 
    
    d = d.set_index(["Name", "Gender"])
    data_list.append(d)
    
data = pd.concat(data_list ,ignore_index = False, axis = 1).fillna(0)

# we don't want it to ignore column names!
#takes long time
data
```

```python
# Ex: merge dataframes so we have counts for 1980 and 1990 sude by side. which name declined in
#popularity the most between 1980 and 1990.

data80 = pd.read_csv("/data/babynames/yob1980.txt", names = ["Name", "Gender", "Count"])

data90 = pd.read_csv("/data/babynames/yob1990.txt", names = ["Name", "Gender", "Count"])

pd.merge(data80, data90) #most obvious thing that comes to mind.
# if u don't specify "on", it will auto match col names
# in both of these dataframes, tabetha f 112 appeared.
```

```python
data_merged = pd.merge(data80,data90, on = ["Name", "Gender"], how = "outer")
```

```python

 
data_merged["Diff"] = data_merged["Count_x"] - data_merged["Count_y"] 
data_merged.sort("Diff")
```

```python
# outer join: gives max possible rows
# inner join: guarantees no naans
#left join: include ppl from left dataframe , but not necessarily right
#right join: makes sure to include ppl from right dataframe, but not left.

#inner join is default and most common join
#pandas merging is similar to SQL joins.

#pandas has its own join func and operates similarly to merge() and concat()

#if u try merge two columns from 2 dataframes that share same name, then suffix x and y is appended.

```
