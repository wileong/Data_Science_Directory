```python
#ALL you need to know:

#specify what data u want from server THROUGH THE URL
#data is returned, usually in JSON or XML format.

#Ex ProPublica

```

```python
url = "https://projects.propublica.org/nonprofits/api/v1/search.json?order=revenue&page=0&sort_order=desc"
```

```python
import urllib.request as r
```

```python
request = r.urlopen(url)
```

```python
raw_json = request.read()
```

```python

#convert from byte to string before loads
raw_json.decode()[:1000]
#all of a sudden empty because I have already READ IT ONCE; can only call read() once...fixed
```

```python
import json

#must decode first!
data = json.loads(raw_json.decode())
```

```python
data
```

```python
for company in data['filings']:
    org = company['organization']
    print(org['name'], org['income_amount'], org['revenue_amount'])
```
