The following code reads in the text of _War and Peace_, by Leo Tolstoy, and stores the string in a variable called "text".

```python
with open("/data/warandpeace.txt", "r") as f:
    text = f.read()
```

Do not try to print the entire text! Just print the first 1000 characters. (You can treat a string in Python as a list of characters.)

```python
print(text[:1000])
```

To make it easy to count words, let's convert this string into a list of words. Use the .split() function (refer to the Python documentation [here](https://docs.python.org/3/library/stdtypes.html#str.split)).

```python
word_list = text.split() #split by whitespace
word_list
```

Now write a for loop to iterate over the words in word_list and store the counts in a dict called word_counts.

```python
word_counts = {} #empty dict

for word in word_list:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1
        
print(word_counts)
```
